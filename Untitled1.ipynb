```python
import pandas as pd

# Load data from CSV files (keep these variable names unchanged)
tomatoes = pd.read_csv("/content/drive/MyDrive/Time Series Forecasting Class - Fall 2024/exam/data/Tomatoes.csv")
sorghum = pd.read_csv("/content/drive/MyDrive/Time Series Forecasting Class - Fall 2024/exam/data/Sorghum.csv")
potatoes_irish = pd.read_csv("/content/drive/MyDrive/Time Series Forecasting Class - Fall 2024/exam/data/Potatoes (Irish).csv")
peas_fresh = pd.read_csv("/content/drive/MyDrive/Time Series Forecasting Class - Fall 2024/exam/data/Peas (fresh).csv")
oranges_big_size = pd.read_csv("/content/drive/MyDrive/Time Series Forecasting Class - Fall 2024/exam/data/Oranges (big size).csv")
maize = pd.read_csv("/content/drive/MyDrive/Time Series Forecasting Class - Fall 2024/exam/data/Maize.csv")
chili_red = pd.read_csv("/content/drive/MyDrive/Time Series Forecasting Class - Fall 2024/exam/data/Chili (red).csv")
cassava = pd.read_csv("/content/drive/MyDrive/Time Series Forecasting Class - Fall 2024/exam/data/Cassava.csv")
beans_dry = pd.read_csv("/content/drive/MyDrive/Time Series Forecasting Class - Fall 2024/exam/data/Beans (dry).csv")

```

```python
# Define a function to filter relevant columns
def filter_columns(df):
    return df[['cm_name', 'mkt_name', 'mp_year', 'mp_month', 'mp_price']]

# Filter columns for each dataset
a = filter_columns(tomatoes)
b = filter_columns(sorghum)
c = filter_columns(potatoes_irish)
d = filter_columns(peas_fresh)
e = filter_columns(oranges_big_size)
f = filter_columns(maize)
g = filter_columns(chili_red)
h = filter_columns(cassava)
i = filter_columns(beans_dry)
```

```python
# Define a function to convert 'mp_year' and 'mp_month' to datetime
def convert_to_datetime(df):
    df['date'] = pd.to_datetime(df['mp_year'].astype(str) + '-' + df['mp_month'].astype(str), format='%Y-%m', errors='coerce')
    return df

# Convert to datetime for each dataset
a = convert_to_datetime(a)
b = convert_to_datetime(b)
c = convert_to_datetime(c)
d = convert_to_datetime(d)
e = convert_to_datetime(e)
f = convert_to_datetime(f)
g = convert_to_datetime(g)
h = convert_to_datetime(h)
i = convert_to_datetime(i)
```

```python
# Remove 'mp_year' and 'mp_month' columns
def drop_columns(df):
    return df.drop(['mp_year', 'mp_month'], axis=1)

a = drop_columns(a)
b = drop_columns(b)
c = drop_columns(c)
d = drop_columns(d)
e = drop_columns(e)
f = drop_columns(f)
g = drop_columns(g)
h = drop_columns(h)
i = drop_columns(i)
```

```python

# Merge 'cm_name' and 'mkt_name' into a single 'location' column
def merge_names(df):
    df['location'] = df['cm_name'] + ', ' + df['mkt_name']
    return df.drop(['cm_name', 'mkt_name'], axis=1)

a = merge_names(a)
b = merge_names(b)
c = merge_names(c)
d = merge_names(d)
e = merge_names(e)
f = merge_names(f)
g = merge_names(g)
h = merge_names(h)
i = merge_names(i)
```

```python
# Set 'date' as the index for all DataFrames
def set_date_as_index(df):
    return df.set_index('date')

a = set_date_as_index(a)
b = set_date_as_index(b)
c = set_date_as_index(c)
d = set_date_as_index(d)
e = set_date_as_index(e)
f = set_date_as_index(f)
g = set_date_as_index(g)
h = set_date_as_index(h)
i = set_date_as_index(i)
```

```python
df = pd.concat([a, b, c, d, e, f, g, h, i])
```

```python
print(df.head(300))
```

```python
# Combine all dataframes
df_all = pd.concat([a, b, c, d, e, f, g, h, i])

# Reset the index to make 'date' a regular column
df_all = df_all.reset_index()

# Group by 'date' and 'location' and aggregate 'mp_price' (e.g., using mean)
df_all = df_all.groupby(['date', 'location'])['mp_price'].mean().reset_index()

# Create a date range
date_range = pd.date_range(start=df_all['date'].min(), end=df_all['date'].max(), freq='MS')

# Create a new DataFrame with the date range
df_date_range = pd.DataFrame({'date': date_range})

# Merge the aggregated DataFrame with the date range DataFrame
df_all = pd.merge(df_date_range, df_all, on='date', how='left')

# Set 'date' as the index again
df_all = df_all.set_index('date')

print(df_all.head(300))
```

```python
# Filter for 'Tomatoes' only
tomatoes_data = df[df["location"].str.contains("Tomatoes")]

# Ensure the 'date' column is set as the index and sorted
tomatoes_data.index = pd.to_datetime(tomatoes_data.index)
tomatoes_data = tomatoes_data.sort_index()

# Extract the price time series
tomatoes_ts = tomatoes_data["mp_price"]

```

```python
# Calculate moving averages for 3 and 6 months
ma_3 = tomatoes_ts.rolling(window=3).mean()
ma_6 = tomatoes_ts.rolling(window=6).mean()

# Plot moving averages
plt.figure(figsize=(12, 6))
plt.plot(tomatoes_ts, label="Original Time Series", alpha=0.6)
plt.plot(ma_3, label="3-Month Moving Average", linestyle="--")
plt.plot(ma_6, label="6-Month Moving Average", linestyle="--")
plt.title("Moving Averages for Tomatoes")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
# Holt-Winters Exponential Smoothing
from statsmodels.tsa.holtwinters import ExponentialSmoothing

hw_model = ExponentialSmoothing(tomatoes_ts, trend="add", seasonal="add", seasonal_periods=12)
hw_fitted = hw_model.fit()

# Forecast the next 6 months
hw_forecast = hw_fitted.forecast(6)

# Plot the forecast
plt.figure(figsize=(12, 6))
plt.plot(tomatoes_ts, label="Original Time Series")
plt.plot(hw_fitted.fittedvalues, label="Fitted Values (Holt-Winters)", linestyle="--")
plt.plot(hw_forecast.index, hw_forecast, label="Holt-Winters Forecast", linestyle="--")
plt.title("Holt-Winters Exponential Smoothing Forecast for Tomatoes")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()

print("Forecasted next 6 months (Holt-Winters):")
print(hw_forecast)
```

```python
# Apply Holt-Winters Exponential Smoothing
hw_model = ExponentialSmoothing(tomatoes_ts, trend="add", seasonal="add", seasonal_periods=12)
hw_fitted = hw_model.fit()

# Forecast the next 6 months
hw_forecast = hw_fitted.forecast(6)

# Filter the original time series and fitted values to start from 2012
filtered_tomatoes_ts = tomatoes_ts[tomatoes_ts.index >= "2012"]
filtered_fitted_values = hw_fitted.fittedvalues[hw_fitted.fittedvalues.index >= "2012"]

# Plot the results starting from 2012
plt.figure(figsize=(12, 6))
plt.plot(filtered_tomatoes_ts, label="Original Time Series (2012 onward)", alpha=0.6)
plt.plot(filtered_fitted_values, label="Fitted Values (Holt-Winters)", linestyle="--", color="orange")

# Add the forecast (manually adjust index for proper plotting)
forecast_dates = pd.date_range(start=tomatoes_ts.index[-1], periods=6, freq='M')
plt.plot(forecast_dates, hw_forecast, label="Holt-Winters Forecast", linestyle="--", color="green")

# Customize the plot
plt.title("Holt-Winters Exponential Smoothing Forecast (2012 onward)")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()

```

```python
from prophet import Prophet
import matplotlib.pyplot as plt

# Prepare data for Prophet
prophet_data = tomatoes_ts.reset_index()
prophet_data.columns = ['ds', 'y']  # Prophet requires 'ds' and 'y' as column names
prophet_data = prophet_data[prophet_data['ds'] >= '2012']  # Filter data from 2012

# Fit the Prophet model
model = Prophet()
model.fit(prophet_data)

# Make a forecast for the next 6 months
future = model.make_future_dataframe(periods=6, freq='M')
forecast = model.predict(future)

# Plot the forecast
fig = model.plot(forecast)
plt.title("Facebook Prophet Forecast for Tomatoes")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()

# Optional: Plot components (trend, seasonality, etc.)
fig_components = model.plot_components(forecast)
plt.show()

# Display forecasted values
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(6))
```
